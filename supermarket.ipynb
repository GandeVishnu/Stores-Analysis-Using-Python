```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/ADMIN/OneDrive/Desktop/New folder (2)/supermarket.csv")

# Convert the InvoiceDate column to a datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group data by InvoiceDate and calculate the sum of Total
grouped_data = df.groupby('Date')['Total'].sum()

# Create the time series graph
plt.plot(grouped_data.index, grouped_data.values)

plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Time Series Graph of Total Sales')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/ADMIN/OneDrive/Desktop/New folder (2)/supermarket.csv")

# Convert the InvoiceDate column to a datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group data by InvoiceDate and calculate the sum of Total
grouped_data = df.groupby('Date')['Total'].sum()

# Create the area plot
plt.figure(figsize=(12, 6))
plt.fill_between(grouped_data.index, grouped_data.values, alpha=0.5)

plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Time Series Area Plot of Total Sales')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("C:/Users/ADMIN/OneDrive/Desktop/New folder (2)/supermarket.csv")

# KPI 1: Total Revenue
total_revenue = df['Total'].sum()
print(f"Total Revenue: ${total_revenue:.2f}")

# KPI 2: Average Rating
average_rating = df['Rating'].mean()
print(f"Average Rating: {average_rating:.2f}")

# KPI 3: Total Quantity Sold
total_quantity_sold = df['Quantity'].sum()
print(f"Total Quantity Sold: {total_quantity_sold}")

# KPI 4: Gross Margin Percentage
gross_margin_percentage = df['gross margin percentage'].mean()
print(f"Gross Margin Percentage: {gross_margin_percentage:.2f}%")

# KPI 5: Customer Satisfaction (Assuming higher rating indicates higher satisfaction)
customer_satisfaction = df['Rating'].mean()
print(f"Customer Satisfaction: {customer_satisfaction:.2f}")

# KPI 6: Average Transaction Value
average_transaction_value = df['Total'].mean()
print(f"Average Transaction Value: ${average_transaction_value:.2f}")

# KPI 7: Number of Transactions
number_of_transactions = df['Invoice ID'].nunique()
print(f"Number of Transactions: {number_of_transactions}")

# KPI 8: Revenue per Customer
revenue_per_customer = total_revenue / number_of_transactions
print(f"Revenue per Customer: ${revenue_per_customer:.2f}")

# KPI 9: Units Sold per Transaction
units_sold_per_transaction = total_quantity_sold / number_of_transactions
print(f"Units Sold per Transaction: {units_sold_per_transaction:.2f}")

# KPI 10: Average Gross Income
average_gross_income = df['gross income'].mean()
print(f"Average Gross Income: ${average_gross_income:.2f}")

```
